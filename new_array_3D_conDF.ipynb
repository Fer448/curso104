```python
import numpy as np
import pandas as pd

# Par√°metros
num_pacientes = 5
num_fechas = 7

# Generar matriz de valores aleatorios entre 35 y 45
np.random.seed(123)  # Para reproducibilidad
datos = np.random.uniform(35, 45, size=(num_fechas, num_pacientes))

# Crear DataFrame con nombres
fechas = [f"Fecha {i+1}" for i in range(num_fechas)]
pacientes = [f"Paciente {j+1}" for j in range(num_pacientes)]
df = pd.DataFrame(datos, index=fechas, columns=pacientes)

# Mostrar tabla
print(df)
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# 1. Crear datos simulados
np.random.seed(123)
num_pacientes = 5
num_fechas = 7

# Crear DataFrame de temperaturas aleatorias entre 35 y 45
datos = np.random.uniform(35, 45, size=(num_fechas, num_pacientes))
fechas = [f"Fecha {i+1}" for i in range(num_fechas)]
pacientes = [f"Paciente {j+1}" for j in range(num_pacientes)]

df = pd.DataFrame(datos, index=fechas, columns=pacientes)

# 2. Graficar evoluci√≥n de temperatura por paciente
plt.figure(figsize=(10, 6))

for paciente in df.columns:
    plt.plot(df.index, df[paciente], marker='o', label=paciente)

plt.title("Evoluci√≥n de la temperatura por paciente")
plt.xlabel("Fecha")
plt.ylabel("Temperatura (¬∞C)")
plt.ylim(34, 46)
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import pandas as pd

# Datos de ejemplo
np.random.seed(123)
num_pacientes = 5
num_fechas = 7

datos = np.random.uniform(35, 45, size=(num_fechas, num_pacientes))
fechas = [f"Fecha {i+1}" for i in range(num_fechas)]
pacientes = [f"Paciente {j+1}" for j in range(num_pacientes)]

df = pd.DataFrame(datos, index=fechas, columns=pacientes)

# Crear la fila con nombres (como una serie)
fila_nombres = pd.DataFrame([pacientes], index=["Nombres"], columns=pacientes)

# Convertir df a string para que la fila con texto pueda unirse
df_str = df.astype(str)

# A√±adir la fila al principio
df_con_encabezado = pd.concat([fila_nombres, df_str])

# Mostrar resultado
print(df_con_encabezado)
```

```python
import numpy as np
import pandas as pd

# Datos de ejemplo
np.random.seed(123)
num_pacientes = 5
num_fechas = 7

datos = np.random.uniform(35, 45, size=(num_fechas, num_pacientes))
fechas = [f"Fecha {i+1}" for i in range(num_fechas)]
pacientes = [f"Paciente {j+1}" for j in range(num_pacientes)]

df = pd.DataFrame(datos, index=fechas, columns=pacientes)

# A√±adir columna de fechas al principio
df_con_fecha = df.copy()
df_con_fecha.insert(0, "Fecha", df_con_fecha.index)

# Mostrar
print(df_con_fecha)
```

```python
# Intercambiar filas 2 y 3 (√≠ndices 1 y 2 porque Python es base 0)
df_copy = df_con_fecha.copy()
df_copy.iloc[[1, 2]] = df_copy.iloc[[2, 1]].values

# Mostrar tabla
print(df_copy)
```

```python
# Ordenar alfab√©ticamente por la columna "Fecha"
df_ordenado = df_con_fecha.sort_values(by="Fecha")

# Mostrar tabla ordenada
print(df_ordenado)
```
