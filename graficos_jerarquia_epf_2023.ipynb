```python
import pandas as pd
import plotly.express as px

# Cargar el archivo Excel
file_path = "/content/epf_spain_cmadrid.xlsx"
df = pd.read_excel(file_path)

# Limpiar datos: eliminar filas vacías
df_clean = df.dropna(subset=["Porcentaje de gasto 2023", "Total Nacional", "13 Madrid, Comunidad de", "Madrid, Ciudad"])

# Columnas con los valores a graficar
value_columns = ["Total Nacional", "13 Madrid, Comunidad de", "Madrid, Ciudad"]

# Crear y mostrar gráficos sunburst
for col in value_columns:
    fig = px.sunburst(
        df_clean,
        path=["Porcentaje de gasto 2023"],
        values=col,
        title=f"Distribución del gasto - {col}"
    )
    fig.show()
```

```python
import pandas as pd
import plotly.express as px

# Cargar el archivo Excel
file_path = "/content/epf_spain_cmadrid.xlsx"
df = pd.read_excel(file_path)

# Limpiar los datos: eliminar filas vacías
df_clean = df.dropna(subset=["Porcentaje de gasto 2023", "Total Nacional", "13 Madrid, Comunidad de", "Madrid, Ciudad"])

# Columnas con valores numéricos para graficar
value_columns = ["Total Nacional", "13 Madrid, Comunidad de", "Madrid, Ciudad"]

# Crear un treemap para cada columna
for col in value_columns:
    fig = px.treemap(
        df_clean,
        path=["Porcentaje de gasto 2023"],
        values=col,
        title=f"Distribución del gasto - {col}",
    )
    fig.show()
```
