```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression

# Cargar los datos desde el archivo CSV
df = pd.read_csv(r"C:\Users\usuario2\Desktop\curso 104\Python\ejemplos python\Ejemplos\Ejemplos\children_height_age.csv")

# Mostrar el DataFrame completo (o las primeras filas si es muy largo)
print("Datos cargados:")
print(df.head())  # Usa df si quieres mostrar todo

# Seleccionar la variable independiente (edad) y la dependiente (altura)
X = df[["age"]]
y = df["height"]

# Crear y ajustar el modelo de regresi√≥n lineal
model = LinearRegression()
model.fit(X, y)

# Imprimir los coeficientes del modelo
print(f"\nModelo de regresi√≥n:")
print(f"Intercepto: {model.intercept_:.2f}")
print(f"Coeficiente de edad: {model.coef_[0]:.2f}")

# Visualizar los datos y la l√≠nea de regresi√≥n
plt.figure(figsize=(8, 5))
sns.scatterplot(x="age", y="height", data=df, hue="sex", palette="Set1")
plt.plot(df["age"], model.predict(X), color="black", linewidth=2, label="Regresi√≥n lineal")
plt.xlabel("Edad (a√±os)")
plt.ylabel("Altura (cm)")
plt.title("Regresi√≥n lineal: Altura vs Edad")
plt.legend()
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression

# Cargar los datos
df = pd.read_csv(r"C:\Users\usuario2\Desktop\curso 104\Python\ejemplos python\Ejemplos\Ejemplos\children_height_age.csv")

# Mostrar las primeras filas del DataFrame
print("Datos cargados:")
print(df.head())

# Separar por sexo
df_male = df[df["sex"] == "male"]
df_female = df[df["sex"] == "female"]

# Modelo para hombres
X_male = df_male[["age"]]
y_male = df_male["height"]
model_male = LinearRegression().fit(X_male, y_male)

# Modelo para mujeres
X_female = df_female[["age"]]
y_female = df_female["height"]
model_female = LinearRegression().fit(X_female, y_female)

# Mostrar coeficientes
print("\nModelo para hombres:")
print(f"  Intercepto: {model_male.intercept_:.2f}")
print(f"  Coeficiente de edad: {model_male.coef_[0]:.2f}")

print("\nModelo para mujeres:")
print(f"  Intercepto: {model_female.intercept_:.2f}")
print(f"  Coeficiente de edad: {model_female.coef_[0]:.2f}")

# Visualizar los datos y modelos
plt.figure(figsize=(8, 5))
sns.scatterplot(x="age", y="height", data=df, hue="sex", palette="Set1")

# L√≠neas de regresi√≥n
plt.plot(X_male, model_male.predict(X_male), color="blue", label="Regresi√≥n hombres")
plt.plot(X_female, model_female.predict(X_female), color="red", label="Regresi√≥n mujeres")

plt.xlabel("Edad (a√±os)")
plt.ylabel("Altura (cm)")
plt.title("Regresi√≥n lineal por sexo: Altura vs Edad")
plt.legend()
plt.tight_layout()
plt.show()
```
